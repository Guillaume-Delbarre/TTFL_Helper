```python
from nba_api.stats.endpoints import CommonAllPlayers

SEASON = '2025-26'

players_df = CommonAllPlayers(is_only_current_season=1).get_data_frames()[0]
players_ids = players_df["PERSON_ID"].unique()
```

```python
players_df
```

```python
from tqdm import tqdm
from nba_api.stats.endpoints import PlayerGameLog
import time

logs = []

for pid in tqdm(players_ids[:10]) :
    try :
        game_log = PlayerGameLog(player_id=pid, season=SEASON).get_data_frames()[0]
        if not game_log.empty :
            logs.append(game_log)
    except Exception as e :
        print(f"[ERROR] Exception : {e}")
    time.sleep(0.6)  # throttle to avoid rate limits
```
